## 회귀분석의 최소제곱 추정

diabetes 설명

당뇨 환자 442명 db

종속 변수 : 당뇨 치료 1년 후 병세 호전 정도

설명 변수 : 나이, 성별, 체질량지수, 평균혈압, 혈청 측정값

```r
data(diabetes)
ols = solve(t(diabetes$x)%*%diabetes$x)%*%t(diabetes$x)%*%diabetes$y
ols
```

## 능형회귀분석 회귀계수의 추정

```r
library(glmnet)
r.rslt = glmnet(x=diabetes$x, y=diabetes$y, family="gaussian", alpha=0) # alpha를 1로 하면 라쏘회귀분석이 된다.
r.rslt$beta
```

## 라쏘회귀분석 회귀계수의 추정

```r
r.rslt = glmnet(x=diabetes$x, y=diabetes$y, family="gaussian", alpha=1)
r.rslt$beta
```

## 빅데이터 환경에서의 회귀분석

diabetes가 빅데이터 데이터는 아니지만, 여기선 빅데이터라고 간주하고 실습하기로 한다.

```r
install.packages('biglm')
install.packages('bigmemory')
install.packages('biganalytics')
```

```r
library(biglm)
library(bigmemory)
library(biganalytics)
```

## 빅데이터 환경에서 회귀분석하기

```r
big.data = as.big.matrix(as.matrix(diabetes))
fm = y~x.age+x.sex+x.bmi+x.map+x.tc+x.ldl+x.hdl+x.tch+x.ltg+x.glu
lm.rslt = biglm.big.matrix(fm,big.data)
summary(lm.rslt)
```

## 빅데이터 환경에서 라쏘회귀분석

```r
install.packages('ff')
```

```r
library(ff)
```

```r
install.packages('biglars')
```

```r
# as.ff 함수는 객체를 ff 객체로 바꿔주는 함수이다.
library(biglars)
ff.x = as.ff(diabetes$x)
ff.y = as.ff(diabetes$y)
big.lasso = biglars.fit(ff.x, ff.y, type="lasso")

# “package ‘biglars’ is not available for this version of R

big.lasso$coef
big.lasso$RSS
```
